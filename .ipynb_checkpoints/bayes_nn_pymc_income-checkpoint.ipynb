```python
def construct_nn(ann_input, ann_output):
    n_hidden = 20

    numFeatures = 4;

    init_in_1 = np.random.randn(numFeatures, n_hidden).astype(floatX)
    init_1_2 = np.random.randn(n_hidden, n_hidden).astype(floatX)
    init_2_out = np.random.randn(n_hidden, 1).astype(floatX)

    with pm.Model() as neural_network:
        weights_in_1 = pm.Normal('w_in_1', 0, sd=1,
                                 shape=(numFeatures, n_hidden), testval=init_in_1)
        weights_1_2 = pm.Normal('w_1_2', 0, sd=1,
                                shape=(n_hidden, n_hidden), testval=init_1_2)
        weights_2_out = pm.Normal('w_2_out', 0, sd=1,
                                  shape=(n_hidden,1), testval=init_2_out)

        act_1 = pm.math.tanh(pm.math.dot(ann_input, weights_in_1))
        act_2 = pm.math.tanh(pm.math.dot(act_1, weights_1_2))
        act_3 = pm.math.dot(act_2, weights_2_out)
        
        out = pm.Normal('out', mu=act_3, sd=10., observed=ann_output)

    return neural_network

ann_input = theano.shared(X_train.as_matrix())
ann_output = theano.shared(Y_train.as_matrix())
neural_network = construct_nn(ann_input, ann_output)
```

```python
from pymc3.theanof import set_tt_rng, MRG_RandomStreams
set_tt_rng(MRG_RandomStreams(42))

%time
with neural_network:
    trace = pm.sample(draws=100)
```

```python
with neural_network:
    pm.traceplot(trace)
```

```python
with neural_network:
    ppc = pm.sample_ppc(trace, 500)
    rpred = ppc['out'].mean(axis=0)
    fig, ax = plt.subplots()
    sns.distplot(rpred)
```

```python
ann_input.set_value(X_test.as_matrix())
#ann_output.set_value(Y_test['avg_total_income'])

with neural_network:    
    ppc = pm.sample_ppc(trace, 500)
    pred = ppc['out'].mean(axis=0)
    proba = ppc['out'].std(axis=0)

plt.figure()
ax = sns.distplot(ppc['out'][0], label='Posterior predictive means')
#ax.axvline(Y_train.mean(), color='r', ls='--', label='True mean')
ax.legend()
#print('Accuracy: {}'.format( Y_test - ppc['out'])
```

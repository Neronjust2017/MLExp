```python
#*********************@Author: Aniruddh Khera*********************
# About the data: S&P500 companies stock prices from 2010-2016
# Here I analyzed for GOOGLE, we can do it for other
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb  
import numpy as np
import pymc3 as pm
from pymc3.distributions.timeseries import *
import theano
floatX = theano.config.floatX
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("prices-split-adjusted.csv")
```

```python
#**********************EXPLORATORY ANALYSIS**********************
#select google
goog = df[df.symbol == 'GOOG']
goog.date = goog.date.convert_objects(convert_dates='coerce')
goog = goog.sort_index(by="date")
goog = goog.set_index("date")
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(goog['close'], label='Google')
ax.set(xlabel='time', ylabel='Close')
ax.legend();
```

```python
# Split data into training & testing
goog_train = goog[goog.index<'2016-12-15']
goog_test = goog[goog.index>='2016-12-15']

# Take log to make variance constant
goog_train['Log close'] = goog_train['close'].apply(lambda x: np.log(x))  
goog_train['Log open'] = goog_train['close'].apply(lambda x: np.log(x))  
goog_test['Log close'] = goog_test['close'].apply(lambda x: np.log(x))  
goog_test['Log open'] = goog_test['close'].apply(lambda x: np.log(x))  
```

```python
# Trying to emulate http://twiecki.github.io/blog/2017/03/14/random-walk-deep-net/
def construct_nn(ann_input, ann_output):   
    n_hidden =  5

    # Initialize random weights between each layer
    init_1 = np.random.randn(goog_train['Log open'].shape[0], n_hidden).astype(floatX)
    init_2 = np.random.randn(n_hidden, n_hidden).astype(floatX)
    init_out = np.random.randn(n_hidden).astype(floatX)   
    with pm.Model() as neural_network:
    
        # Weights from input to hidden layer   
        step_size = pm.HalfNormal('step_size', sd=np.ones(n_hidden), 
                              shape=n_hidden)
        weights_in_1 = pm.Normal('w1', sd=step_size,
                           shape=(goog_train['Log open'].shape[0],n_hidden), testval=init_1)

        weights_1_2 = pm.Normal('w2', mu=0, sd=1., 
                            shape=(1, n_hidden, n_hidden),
                            testval=init_2)
       
        weights_2_out = pm.Normal('w3', mu=0, sd=1.,
                              shape=(n_hidden,),
                              testval=init_out)
     

        # Build neural-network using tanh activation function
        act_1 =  pm.math.tanh(pm.math.dot(ann_input, weights_in_1))
        act_2 =  pm.math.tanh(pm.math.dot(act_1, weights_1_2))
        act_out = pm.math.sigmoid(pm.math.dot(act_2,weights_2_out))
        out = pm.Normal('out', act_out, observed=ann_output, total_size=goog_train['Log close'].shape[0])
    return neural_network

ann_input = theano.shared(np.array(goog_train['Log open']))
ann_output = theano.shared(np.array(goog_train['Log close']))
neural_network = construct_nn(ann_input, ann_output)
```

```python
# http://docs.pymc.io/notebooks/api_quickstart
from pymc3.theanof import set_tt_rng, MRG_RandomStreams
set_tt_rng(MRG_RandomStreams(42))

%time
with neural_network:
    inference = pm.ADVI()
    approx = pm.fit(n=50000, method=inference)
```

```python
# What does the Average loss tells you?
# https://discourse.pymc.io/t/average-loss-in-optimization-output/372/11
with neural_network:
    trace = approx.sample(draws=5000)
    pm.traceplot(trace)
```
